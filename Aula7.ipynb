## 1 Do que se trata uma linguagem ORIENTADA A OBJETOS

Antes de darmos incio ao entendimento do paradigma de programao orientada a objetos (POO),  necessrio que se saiba que existem vrios outros paradigmas. Esses so meios de se classificar uma linguagem em termos de suas funcionalidades. Entre eles podemos citar o paradigma procedural, o funcional, o estruturado, o imperativo e o orientado a objetos.

No mundo do desenvolvimento de softwares existem as linguagens conhecidas como multiparadigmas. Como o nome sugere, essas linguagens suportam mais de um paradigma como  o caso do Python. Um dos paradigmas presentes em Python  a orientao a objetos que iremos abordar nesta aula. 

At agora em nossas aulas trabalhamos com o paradigma procedural, cuja a idia principal consiste na criao de funes (procedimentos) que operam sobre os dados. J na orientao a objetos, o foco muda para a criao de "objetos" que contm atributos e mtodos,  que sero explicados abaixo.

### 1.1 Os quatro pilares que seguem a POO

Existem 4 aspectos que tornam uma linguagem orientada a objetos: encapsulamento, abstrao, herana e polimorfismo. Nesta aula deixaremos herana e polimorfismo como contedo extra.

#### 1.1.1 Abstrao

Precisamos lembrar que estamos querendo trabalhar com conceitos reais de objetos, ento precisamos estudar e entender o que tal representao de um objeto real ir realizar dentro do nosso sistema. <br>

 muito importante identificar o objeto e dar uma indentidade a ele, isso quer dizer que voc entendeu a existncia de um objeto essncial para seu sistema/programa e o definiu de forma que sempre saber quem e o que .  de suma gravidade que a identidade do objeto permanea nica, isso evita conflitos dentro do sistema.<br>

Agora que voc identificou esse objeto,  necessrio saber suas caracteristicas, o que define esse objeto, pelo que ele  constituido? Por exemplo, voc tem um martelo azul, preo de 23 reais e tamanho de 40cm, est claro que as caracteristicas do martelo  azul, tamanho de 40cm e preo de 23 reais. Para ficar mais claro ainda, imagine um cachorro, quais so as possveis caracteristicas? Raa, tamanho, pelo, cor, nome, idade... <br>

Pronto, j sabemos quais caracteristicas nosso objeto possu, precisamos agora saber o que ele realiza, quais aes pelo objeto podem ser tomadas? Essas aes podem ser um latido, se nosso objeto for um cachorro, martelar se nosso objeto for um martelo, cortar se o nosso objeto for uma faca, enfim inmeras aes que dependem da identidade de um objeto.<br>

Obs: Caracteristicas e aes, so vistas como propriedades e mtodos respectivamente em linguagens de POO.


#### 1.1.2 Encapsulamento

Na programao orientada a objetos  possvel omitir certas informaes, como as propriedades dos objetos. Dessa forma,  possvel construir um sistema de forma mais segura  como se envolvessemos essas informaes em caixas pretas, chamamos isso de encapsulamento. 

A funo do encapsulamento  impedir o acesso direto aos atributos do objeto. Para que isso seja possvel, em geral, criam-se mtodos *getters* e *setters*, que retornam e alteram os valores dos atributos.<br>

Talvez o exemplo a seguir facilite o entendimento desse conceito: Imagine que voc est em um caixa eletrnico. Voc insere o seu carto e digita sua senha, dessa forma voc ter acesso a sua conta. Para simplificar, vamos focar apenas no saldo, voc consegue inserir dinheiro (depositar), extrair (sacar) e visulizar o saldo. Podemos afirmar que o usurio no possui acesso direto as informaes da conta. Por exemplo, voc no poderia mudar o saldo da conta sem depositar ou sacar algum valor. 

Fazendo uma analogia com POO, a ao de depositar seria um mtodo modificador(setter) e, a ao de consultar o saldo seria um mtodo consultor (getter).<br>

#### 1.1.3 Famlia - Herana

Imagine que voc est implementando o mtodo de acelerao de muitos carros de diferentes modelos. Voc teria que fazer um cdigo para cada carro. Seria muito mais interessante se todos o modelos pudessem "herdar" um comportamento em comum do conceito de carro. Todo carro acelera, ento poderiamos fazer um nico cdigo para todos os modelos.

Portanto, quando vamos trabalhar com objetos de caracteristicas semelhantes e queremos otimizar isso para evitar linhas de cdigo e aes redundantes usamos o conceito de herana.




#### 1.1.4 Polimorfismo

Esse aspecto est abraado com a herena, pois na POO existir momentos em que objetos tero a adaptar mtodos herdados para que haja uma implementao especfica.  

Por exemplo, um microondas e uma televiso so eletrodomsticos. Se a implementao desse exemplo fosse realizada, a classe eletrodomsico seria a classe pai, enquanto as classes televiso e microondas seriam as classes filhas.

Como todo eletrodomstico liga, esse mtodo precisa estar listado na classe pai. Por conseguinte cada classe filha tambm possuir esse mtodo devido a herana. Porm, como cada aparelho liga de forma distinta, esse mtodo precisa ser reescrito em cada classe filha para que funcione corretamente para cada uma delas.

#### 2.2.4 Declarando atributos e criando um construtor

Normalmente, ao criarmos uma classe, a primeira coisa que escrevemos so os atributos dessa classe. Estes precisaro de um valor ao serem declarados. Por padro, normalmente utilizamos a palavra reservada `None`, que representa um valor nulo. Contudo, isso no  uma regra. Voc pode utilizar `0`, `""` (uma string vazia), ou qualquer outra coisa que faa sentido para a sua aplicao.

Na linguagem Python, existem alguns nomes de mtodos reservados para a prpria linguagem. Um deles  o que chamamos de construtor, que vai inicializar um objeto. Seu nome ser `__init__()`, e ele, assim como todos os mtodos da classe, precisar de um parmetro que representa o prprio objeto, que por conveno, chamamos de <b>self</b>. Alm disso, colocaremos os atributos que queremos armazenar ao criar uma nova conta. Observe:

```python
#Declaramos aqui a nossa classe
class Conta:
    #aqui colocaremos os atributos que todo objeto da conta precisa ter
    nome = None
    cpf = None
    ID = None
    saldo = 0.0
    limite = 0.0
    
    #agora definiremos o construtor da classe
    def __init__(self, nome, cpf, ID, saldo, limite):
        self.nome = nome
        self.cpf = cpf
        self.ID = ID
        self.saldo = saldo
        self.limite = limite
```

Repare na forma como utilizamos `self.nome_do_atributo = nome_do_atributo`. Isso serve para diferenciar quando nos referimos ao atributo da classe e quando nos referimos a varivel da funo construtora. Num geral, quando escrevemos `self.valor`, estamos nos referindo ao atributo da classe. Repare tambm que caso no estejamos utilizando o mesmo nome para o parmetro do mtodo e o atributo da classe, no precisamos escrever `self.`. Veja o exemplo:

#### 2.2.5 Mtodos

Como vimos, nossa conta precisa no s ter atributos, mas tambm funcionalidades. Essas funcionalidades de um objeto so chamadas de mtodos, e funcionaro de maneira extremamente similar s funes `depositar()`, `sacar()` e `consultar_saldo()`.

Vamos comear criando juntos a funo `depositar()`. Assim como o mtodo construtor `__init__()`, nosso mtodo `depositar()` precisa do parmetro self, alm do valor a ser depositado.

```python
#Declaramos aqui a nossa classe
class Conta:
    #aqui colocaremos os atributos que todo objeto da conta precisa ter
    nome = None
    cpf = None
    ID = None
    saldo = 0.0
    limite = 0.0
    
    #agora definiremos o construtor da classe
    def __init__(self, nome, cpf, ID, saldo, limite):
        self.nome = nome
        self.cpf = cpf
        self.ID = ID
        self.saldo = saldo
        self.limite = limite
        
    def depositar(self, valor):
        self.saldo += valor
```

Para chamar o mtodo `depositar()`, utilizaremos uma sintaxe bem similar a qual usavamos para acessar um atributo: `conta.depositar(valor)`. O interpretador associa o objeto <b>conta</b> ao argumento <b>self</b> do mtodo. Note que no precisamos passar <b>conta</b> como argumento, pois isso  feito automaticamente pela linguagem Python.

### 3 Aplicando o encapsulamento

 possvel notar ao longo da utilizao da nossa classe Conta que ainda  possvel alterar atributos como o saldo da conta. A resposta para como proibir essas alteraes diretas est no encapsulamento. Precisamos definir o que em nossa classe  <b>pblico</b> e o que em nossa classe  <b>privado</b>. Mas primeiro, o que significa dizer que um atributo ou mtodo  privado? E o que significa dizer que um atributo ou mtodo  publico?

Caso um atributo ou mtodo seja definido como <b>pblico</b>, significa que ele ser completamente acessvel fora do escopo da classe. Por padro, o que no  definido como privado  considerado pblico, e por isso conseguamos acessar os atributos de Conta fora do escopo da classe.

Caso um atributo ou mtodo seja definido como <b>privado</b>, significa que ele ser inacessvel fora do escopo da classe. Dessa forma, se tentarmos acessar o valor de um atributo privado fora do escopo da classe, receberemos um erro!

Normalmente, sempre definimos nossos atributos como privados e nossos mtodos como pblicos.

#### 3.1 Como definir algo como privado

Como vimos acima, no  necessrio dizer que um atributo ou mtodo  publico, pois ele  criado por padro como publico. Precisamos de alguma informao adicional somente quando quisermos um atributo ou mtodo seja privado. Para isto, basta acrescentar dois underscores (_) na frente do atributo ou mtodo. Isso far com que ele seja privado.

Note que sempre que voc chamar esse atributo privado em algum lugar, voc deve colocar os underscores antes de se referir a ele. Vamos ver um exemplo:

```python
#Definiremos aqui uma funo que cria um animal genrico
class Animal:
    __nome = None
    __tipo = None
    __som = None
    
    def __init__(self, nome, tipo, som):
        self.__nome = nome
        self.__tipo = tipo
        self.__som = som
        
    def faz_som(self):
        return self.__som
```

```python
#Criamos aqui um objeto do tipo animal
animal = Animal("Gato", "Onvoro", "Miau")
```

```python
#Tenta acessar o nome do animal indevidamente
print(animal.nome)
```

```python
print(animal.faz_som())
```

#### 3.2 Getters e Setters

Como agora sempre tentaremos colocar nossos atributos como privados, como acessaremos (seja para leitura ou para escrita) um atributo? Para isso, costumamos criar mtodos chamados Getters (para pegar alguma informao) e Setters (para alterar alguma informao). Normalmente os Getters seguem por conveno a regra de nomenclatura `get_nome_do_atributo()` enquanto os Setters recebem por conveno a regra de nomenclatura `set_nome_do_atributo()`.

Aproveitando o exemplo da classe Animal, ser criado o primeiro getter e setter para o atributo nome, enquanto os getters e setters para tipo e som ficam como exerccio.

```python
class Animal:
    __nome = None
    __tipo = None
    __som = None
    
    def __init__(self, nome, tipo, som):
        self.__nome = nome
        self.__tipo = tipo
        self.__som = som
        
    def faz_som(self):
        return self.__som
    
    def get_nome(self):
        return self.__nome
    
    def set_nome(self, nome):
        self.__nome = nome
        
    #implemente aqui os outros getters e setters
```
