### Atividade 1 - Problema do Elevador

O elevador que leva caixas até a porta do foguete precisa ter base de 2,63m por 2,29m e altura de 3,45m. Faça uma função que diga se é possível inserir um container de base retangular de dimensões X,XX x Y,YY x H,HH ou um de base circular com dimensões R.RR X J.JJ no elevador, existe também um peso limite de 1200kg para o elevador, verifique se o elevador aguenta o peso.

<b> - Especificações da função: </b> <br>
<b>Parâmetros:</b> dimensões do container em metros, massa do container em quilogramas.<br>
<b>Saída:</b> valor booleano informando se o container pode ser colocado no elevador ou não.

```python
# Crie função aqui:
def ehPossivelInserirContainer(baseA, baseB, altura, massa):
    if(massa > 1200.00):
        return False
    if(altura > 3.45):
        return False
    if((baseA <= 2.63 and baseB <= 3.45) or
       (baseA <= 3.45 and baseB <= 2.63)):
        return True
    else:
        return False
    
# Base circular:
def ehPossivelInserirContainerCircular(raio, altura, massa):
    return ehPossivelInserirContainer(raio, raio, altura, massa)
```

```python
# Testes
print(ehPossivelInserirContainer(5, 2, 2, 230))  # False
print(ehPossivelInserirContainer(3, 2, 2, 200))  # True
print(ehPossivelInserirContainer(3, 2, 2, 1220)) # True

# Testes com base circular
print(ehPossivelInserirContainerCircular(5, 2, 230))   # False
print(ehPossivelInserirContainerCircular(2.32, 2, 200))# True
print(ehPossivelInserirContainerCircular(3, 2, 1220))  # False
```

### Atividade 2 - Problema do Termômetro

A NASA pretende fazer parceria com a AEB para auxiliar no projeto de exploração espacial, entretanto, uma das condições para que isso aconteça é que você implemente um conversor de Fahrenheit para Celsius e vice-versa.<br>

<b> - Especificações da função: </b> <br>
Fahrenheit = (Celsius + 32) * 9 / 5<br>
<b>Parâmetros:</b> Temperatura em Fahrenheit, ou Celsius, dependendo da função.<br>
<b>Saída:</b> Temperatura em Celsius ou Fahrenheit.

```python
# Crie funções aqui:
def FahrenheitParaCelsius(temp):
    return (temp - 32) * 5/9

def CelsiusParaFahrenheit(temp):
    return temp * 9/5 + 32
```

```python
# Testes:
print(FahrenheitParaCelsius(32),"C")     # 0F
print(FahrenheitParaCelsius(212),"C")    # 100F
print(FahrenheitParaCelsius(100),"C")    # 37.78F
print(CelsiusParaFahrenheit(0),"F")      # 32F
print(CelsiusParaFahrenheit(100),"F")    # 212F
print(CelsiusParaFahrenheit(37.77),"F")  # 100F
```
