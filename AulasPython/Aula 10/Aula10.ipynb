# Programao competitiva

A programao  uma habilidade fundamental para um profissional do futuro. O mercado de trabalho nessa rea est se desenvolvendo cada dia mais e, por conseguinte, a demanda de profissionais qualificados aumenta diariamente. Sendo assim,  necessrio que os programadores busquem desenvolver habilidades que muitas vezes no so ensinadas at mesmo nos cursos de nvel superior. 

Uma boa forma de ganhar experincia e desenvolver a capacidade de resolver problemas, so os torneios de programao. De ano para ano temos observado que as instituies e principalmente as grandes empresas da rea tm valorizado os alunos que participam dos torneios mais renomados.

Um torneio de programao  classificado como um mind sport (como o xadrez), e so realizados via internet ou rede local. Normalmente, envolvem estudantes da rea que tentam resolver uma srie de problemas de acordo com algumas especificaes, fazendo o uso da programao. Abaixo explicaremos um pouco sobre os principais torneios de programao.

<a id = 'section1'></a>

## 1. Topcom

O topcom, Torneio de Programao de Computadores,  promovido pelo grupo PET Engenharia de Computao (UFES) e,  aberto a estudantes da rea. A competio ocorre durante um dia inteiro, sendo que no perodo matutino ocorrem palestras de profissionais da rea e logo aps isso, a sesso de aquecimento tem incio. Nela, as equipes resolvem alguns problemas sem valor de competio, isso  feito para que os competidores se acostumem com a plataforma de correo utilizada.

A competio acontece no perodo vespertino, onde as equipes resolvem os problemas propostos pela organizao do evento. A competio  em disputada em equipes de trs pessoas, e apenas um computador  disponibilizado por equipe. Assim, o torneio tambm explora o trabalho em equipe sob presso e estimula a criatividade.

No fim do dia ocorrem as premiaes e uma confraternizao entre as equipes. Esse projeto visa motivar a participao de instituies locais em competies nacionais universitrias, como a Maratona de Programao. A elaborao e validao dos problemas contam com a participao de professores do Departamento de Informtica da UFES. O torneio utiliza o BOCA como plataforma de correo automtica.

## 2. Olimpada Brasileira de Informtica

A Sociedade Brasileira de Computao (SBC) organiza anualmente, desde 1999, a Olimpada Brasileira de Informtica (OBI). O pblico alvo dessa competio so os estudantes do ensino fundamental, mdio e do primeiro ano de graduao.

Ao contrrio de vrios outros torneios de programao, a OBI  realizada individualmente e, possui uma correo no binria. Parecida com a correo do Run.Codes, o cdigo da questo  analisado e o programa  pontuado de acordo com a quantidade de casos de teste corretos. Os resultados so divulgados aps a realizao da prova.

Outra peculiaridade da OBI  que o tempo total de resoluo das questes no importa, a nica coisa que  levada em considerao  a quantidade de desafios corretos.

## 3. Maratona de Programao

Assim como a OBI, a Maratona de Programao tambm  organizada anualmente pela Sociedade Brasileira de Computao. Ela se destina a alunos de cursos de graduao e incio de ps-graduao na rea de Computao e afins (Cincia da Computao, Engenharia de Computao, Sistemas de Informao, Matemtica, etc).

A Maratona  disputada em times, que so compostos por trs estudantes de uma mesma escola.  Uma escola  definida pela instituio de ensino superior e a cidade de onde vm os alunos participantes do time, ou seja, uma mesma universidade pode ter vrias Escolas. Com isso, os estudantes tentaro resolver durante 5 horas o maior nmero possvel dos 10 ou mais problemas que so entregues no incio da competio. Estes estudantes tm  sua disposio apenas um computador e material impresso (livros, listagens, manuais) para vencer a batalha contra o relgio e os problemas propostos.

A correo dos problemas  automtica e binria, ou seja,  preciso que o programa submetido passe em todos os casos de teste realizados pela banca avaliadora. Assim sendo, existe uma penalidade de tempo (+ 20 min) para cada submisso incorreta de um problema. No fim, o time que conseguir resolver o maior nmero de problemas, no menor tempo,  declarado vencedor. De forma anloga ao Topcom, a Maratona de Programao utiliza o BOCA como a plataforma de correo automtica.

### 3.1 Regras

Para o ano de 2019, as regras citadas a seguir foram vlidas. Cada membro, antes dessa competio, pode ter participado de no mximo uma final mundial do ICPC, de no mximo 4 (quatro) regionais sul-americanas do concurso (ou seja, da Maratona de Programao) e deve ter iniciado seus estudos universitrios no ano de 2015 ou anos posteriores (a contar do incio do primeiro curso universitrio do aluno, mesmo que no tenha completado este curso), ou ter nascido em 1996 ou anos posteriores.

Assim, por exemplo, se voc iniciou seu primeiro curso superior em 2018, pode participar, mesmo que tenha nascido em 1994. Outro exemplo, se voc nasceu em 1996 pode participar, mesmo tendo iniciado sua graduao em 2013.

### 3.2 Ambiente de programao

Na edio deste ano (2019) o sistema operacional usado foi o Ubuntu 18.04, e as linguagens disponveis foram: C, C++, Python 2, Python 3, Java e Kotlin. 

Alm disso, os editores de texto disponveis so: geany, vim, gedit, sublime, VScode entre outros. Caso voc prefira utilizar IDEs, as disponveis so o Eclipse, o Pycharm e o Intellij. Para mais informaes acesse o site http://maratona.ime.usp.br/

## 4. Code Jam


O Google Code Jam  uma competio internacional hospedado e administrado pelo Google. A competio comeou em 2003 como um meio de identificar os melhores talentos de engenharia para um emprego potencial no Google. Os competidores podem usar qualquer linguagem de programao e ambiente de desenvolvimento para obter suas solues.

A competio  aberta para qualquer pessoa maior de 18 anos e todas as etapas so online, exceto a final. As etapas afunilam os competidores at que sobrem 25 pessoas para disputarem a final em alguma sede do Google.
Para mais informaes acesse https://codingcompetitions.withgoogle.com/codejam/about/
