## 1 Do que se trata uma linguagem ORIENTADA A OBJETOS

Antes de darmos início ao entendimento do paradigma de programação orientada a objetos (POO), é necessário que se saiba que existem vários outros paradigmas. Esses são meios de se classificar uma linguagem em termos de suas funcionalidades. Entre eles podemos citar o paradigma procedural, o funcional, o estruturado, o imperativo e o orientado a objetos.

No mundo do desenvolvimento de softwares existem as linguagens conhecidas como multiparadigmas. Como o nome sugere, essas linguagens suportam mais de um paradigma como é o caso do Python. Um dos paradigmas presentes em Python é a orientação a objetos que iremos abordar nesta aula. 

Até agora em nossas aulas trabalhamos com o paradigma procedural, cuja a idéia principal consiste na criação de funções (procedimentos) que operam sobre os dados. Já na orientação a objetos, o foco muda para a criação de "objetos" que contém atributos e métodos,  que serão explicados abaixo. 


### IMPORTANTE 


### 1.1 Os quatro pilares que seguem a POO

Existem 4 aspectos que tornam uma linguagem orientada a objetos: encapsulamento, abstração, herança e polimorfismo. Nesta aula deixaremos herança e polimorfismo como conteúdo extra.

#### 1.1.1 Abstração

Precisamos lembrar que estamos querendo trabalhar com conceitos reais de objetos, então precisamos estudar e entender o que tal representação de um objeto real irá realizar dentro do nosso sistema. <br>

É muito importante identificar o objeto e dar uma indentidade a ele, isso quer dizer que você entendeu a existência de um objeto essêncial para seu sistema/programa e o definiu de forma que sempre saberá quem e o que é. É de suma gravidade que a identidade do objeto permaneça única, isso evita conflitos dentro do sistema.<br>

Agora que você identificou esse objeto, é necessário saber suas caracteristicas, o que define esse objeto, pelo que ele é constituido? Por exemplo, vocẽ tem um martelo azul, preço de 23 reais e tamanho de 40cm, está claro que as caracteristicas do martelo é azul, tamanho de 40cm e preço de 23 reais. Para ficar mais claro ainda, imagine um cachorro, quais são as possíveis caracteristicas? Raça, tamanho, pelo, cor, nome, idade... <br>

Pronto, já sabemos quais caracteristicas nosso objeto possuí, precisamos agora saber o que ele realiza, quais ações pelo objeto podem ser tomadas? Essas ações podem ser um latido, se nosso objeto for um cachorro, martelar se nosso objeto for um martelo, cortar se o nosso objeto for uma faca, enfim inúmeras ações que dependem da identidade de um objeto.<br>

Obs: Caracteristicas e ações, são vistas como propriedades e métodos respectivamente em linguagens de POO.


#### 1.1.2 Encapsulamento

Na programação orientada a objetos é possível omitir certas informações, como as propriedades dos objetos. Dessa forma, é possível construir um sistema de forma mais segura é como se envolvessemos essas informações em caixas pretas, chamamos isso de encapsulamento. 

A função do encapsulamento é impedir o acesso direto aos atributos do objeto. Para que isso seja possível, em geral, criam-se métodos *getters* e *setters*, que retornam e alteram os valores dos atributos.<br>

Talvez o exemplo a seguir facilite o entendimento desse conceito: Imagine que você está em um caixa eletrônico. Você insere o seu cartão e digita sua senha, dessa forma você terá acesso a sua conta. Para simplificar, vamos focar apenas no saldo, você consegue inserir dinheiro (depositar), extrair (sacar) e visulizar o saldo. Podemos afirmar que o usuário não possui acesso direto as informações da conta. Por exemplo, você não poderia mudar o saldo da conta sem depositar ou sacar algum valor. 

Fazendo uma analogia com POO, a ação de depositar seria um método modificador(setter) e, a ação de consultar o saldo seria um método consultor (getter).<br>

#### 1.1.3 Família - Herança

Imagine que você está implementando o método de aceleração de muitos carros de diferentes modelos. Você teria que fazer um código para cada carro. Seria muito mais interessante se todos o modelos pudessem "herdar" um comportamento em comum do conceito de carro. Todo carro acelera, então poderiamos fazer um único código para todos os modelos.

Portanto, quando vamos trabalhar com objetos de caracteristicas semelhantes e queremos otimizar isso para evitar linhas de código e ações redundantes usamos o conceito de herança.




#### 1.1.4 Polimorfismo

Esse aspecto está abraçado com a herença, pois na POO existirá momentos em que objetos terão a adaptar métodos herdados para que haja uma implementação específica.  

Por exemplo, um microondas e uma televisão são eletrodomésticos. Se a implementação desse exemplo fosse realizada, a classe eletrodomésico seria a classe pai, enquanto as classes televisão e microondas seriam as classes filhas.

Como todo eletrodoméstico liga, esse método precisa estar listado na classe pai. Por conseguinte cada classe filha também possuirá esse método devido a herança. Porém, como cada aparelho liga de forma distinta, esse método precisa ser reescrito em cada classe filha para que funcione corretamente para cada uma delas.

### 2.2 De forma POO
No modo procedural, conseguimos alterar informações de forma direta, como o nome do titular da conta, seu cpf, entre outras coisas. Antes de vermos como classes e objetos funcionam na prática, vamos entender o que é uma classe e o que é um objeto em um programa orientado a objetos.

Uma analogia bem interessante que podemos fazer é entre o projeto de uma casa (a planta dessa casa) e a casa em si. Podemos associar a casa em si como um <b>objeto</b>, enquanto associamos o projeto da casa com a <b>classe</b>. 

#### 2.2.1 Classes
Uma classe é um conjunto de propriedades, caracteristicas, informações, etc., de alguma coisa, além disso também indentifica o comportamento dessa coisa, suas ações, funções, enfim, é um agrupamento descritivo de algo. <br>

Uma classe pode ser comparada às estruturas (tipos) antes vistas no paradigma procedural. Tendo isso em mente, definimos uma classe como uma estrutura capaz de abstrair um conjunto de objetos com características similares.<br>

Traçamos um paralelo com os objetos que conhecemos na vida real, estes que possuem características próprias que os definem (chamadas em POO de atributos) e também utilidades próprias ou funções (métodos).

#### 2.2.4 Declarando atributos e criando um construtor

Normalmente, ao criarmos uma classe, a primeira coisa que escrevemos são os atributos dessa classe. Estes precisarão de um valor ao serem declarados. Por padrão, normalmente utilizamos a palavra reservada `None`, que representa um valor nulo. Contudo, isso não é uma regra. Você pode utilizar `0`, `""` (uma string vazia), ou qualquer outra coisa que faça sentido para a sua aplicação.

Na linguagem Python, existem alguns nomes de métodos reservados para a própria linguagem. Um deles é o que chamamos de construtor, que vai inicializar um objeto. Seu nome será `__init__()`, e ele, assim como todos os métodos da classe, precisará de um parâmetro que representa o próprio objeto, que por convenção, chamamos de <b>self</b>. Além disso, colocaremos os atributos que queremos armazenar ao criar uma nova conta. Observe:

```python
#Declaramos aqui a nossa classe
class Conta:
    #aqui colocaremos os atributos que todo objeto da conta precisa ter
    nome = None
    cpf = None
    ID = None
    saldo = 0.0
    limite = 0.0
    
    #agora definiremos o construtor da classe
    def __init__(self, nome, cpf, ID, saldo, limite):
        self.nome = nome
        self.cpf = cpf
        self.ID = ID
        self.saldo = saldo
        self.limite = limite
```

Repare na forma como utilizamos `self.nome_do_atributo = nome_do_atributo`. Isso serve para diferenciar quando nos referimos ao atributo da classe e quando nos referimos a variável da função construtora. Num geral, quando escrevemos `self.valor`, estamos nos referindo ao atributo da classe. Repare também que caso não estejamos utilizando o mesmo nome para o parâmetro do método e o atributo da classe, não precisamos escrever `self.`. Veja o exemplo:

```python
class Teste:
    atributo1 = None
    atributo2 = ""
    
    def __init__(self, atr1, atr2):
        atributo1 = atr1
        atributo2 = atr2
```

#### 2.2.5 Métodos

Como vimos, nossa conta precisa não só ter atributos, mas também funcionalidades. Essas funcionalidades de um objeto são chamadas de métodos, e funcionarão de maneira extremamente similar às funções `depositar()`, `sacar()` e `consultar_saldo()`.

Vamos começar criando juntos a função `depositar()`. Assim como o método construtor `__init__()`, nosso método `depositar()` precisa do parâmetro self, além do valor a ser depositado.

```python
#Declaramos aqui a nossa classe
class Conta:
    #aqui colocaremos os atributos que todo objeto da conta precisa ter
    nome = None
    cpf = None
    ID = None
    saldo = 0.0
    limite = 0.0
    
    #agora definiremos o construtor da classe
    def __init__(self, nome, cpf, ID, saldo, limite):
        self.nome = nome
        self.cpf = cpf
        self.ID = ID
        self.saldo = saldo
        self.limite = limite
        
    def depositar(self, valor):
        self.saldo += valor
```

Para chamar o método `depositar()`, utilizaremos uma sintaxe bem similar a qual usavamos para acessar um atributo: `conta.depositar(valor)`. O interpretador associa o objeto <b>conta</b> ao argumento <b>self</b> do método. Note que não precisamos passar <b>conta</b> como argumento, pois isso é feito automaticamente pela linguagem Python.

### 3 Aplicando o encapsulamento

É possível notar ao longo da utilização da nossa classe Conta que ainda é possível alterar atributos como o saldo da conta. A resposta para como proibir essas alterações diretas está no encapsulamento. Precisamos definir o que em nossa classe é <b>público</b> e o que em nossa classe é <b>privado</b>. Mas primeiro, o que significa dizer que um atributo ou método é privado? E o que significa dizer que um atributo ou método é publico?

Caso um atributo ou método seja definido como <b>público</b>, significa que ele será completamente acessível fora do escopo da classe. Por padrão, o que não é definido como privado é considerado público, e por isso conseguíamos acessar os atributos de Conta fora do escopo da classe.

Caso um atributo ou método seja definido como <b>privado</b>, significa que ele será inacessível fora do escopo da classe. Dessa forma, se tentarmos acessar o valor de um atributo privado fora do escopo da classe, receberemos um erro!

Normalmente, sempre definimos nossos atributos como privados e nossos métodos como públicos.

#### 3.1 Como definir algo como privado

Como vimos acima, não é necessário dizer que um atributo ou método é publico, pois ele é criado por padrão como publico. Precisamos de alguma informação adicional somente quando quisermos um atributo ou método seja privado. Para isto, basta acrescentar dois underscores (_) na frente do atributo ou método. Isso fará com que ele seja privado.

Note que sempre que você chamar esse atributo privado em algum lugar, você deve colocar os underscores antes de se referir a ele. Vamos ver um exemplo:

```python
#Definiremos aqui uma função que cria um animal genérico
class Animal:
    __nome = None
    __tipo = None
    __som = None
    
    def __init__(self, nome, tipo, som):
        self.__nome = nome
        self.__tipo = tipo
        self.__som = som
        
    def faz_som(self):
        return self.__som
```

```python
#Criamos aqui um objeto do tipo animal
animal = Animal("Gato", "Onívoro", "Miau")
```

```python
#Tenta acessar o nome do animal indevidamente
print(animal.nome)
```

```python
print(animal.faz_som())
```

#### 3.2 Getters e Setters

Para isso, costumamos criar métodos chamados Getters (para pegar alguma informação) e Setters (para alterar alguma informação). Normalmente os Getters seguem por convenção a regra de nomenclatura `get_nome_do_atributo()` enquanto os Setters recebem por convenção a regra de nomenclatura `set_nome_do_atributo()`.
