## 1- Os operadores de comparao

No mundo da programao, um conceito muito til e verstil  o da comparao. Em Python temos alguns operadores que conseguem realizar comparaoes entre valores. Eles so:

Precisamos de uma maneira de pedir para o computador fazer coisas diferentes baseado na comparao entre os valores. Para isso usamos o termo `if`, que vem do ingls para `se`. Com esse termo, podemos pedir para o computador fazer `alguma coisa` `se` uma `condio` for atendida. Vamos tentar entender melhor:<br>

   Precisamos fazer com  que o computador<br>
    `imprima o valor 'a'` `se` `o valor 'a' for maior que o valor 'b'`<br>

Entendeu? Agora vamos ver como escrever coisas desse tipo em um programa de verdade.<br>


Em Python utilizamos a expresso `if` da seguinte forma:<br>
    `if 'Alguma condio':
        'Alguma ao'`<br>

Perceba a indentao, pois ela  extremamente importante e precisa ser respeitada, assim como os dois pontos.<br>

A condio pode ser qualquer valor, e ser considerada verdadeira para o booleano `True` e qualquer valor diferente de zero, nada (`None`) ou vazio, e falsa para o booleano `False` e qualquer um dos valores citados anteriormente. Vamos testar:


```python
#Vamos considerar a varivel

a = True

#Troque o valor de 'a' e veja os casos em que seu valor  considerado verdadeiro
if a:
    print('A condio  verdadeira')

```

O operador `elif`  uma contrao do termo `else if` (se voc j viu linguagens como C talvez conhea dessa forma) e significa `caso contrrio, se`. Ele  muito til para checar mais de uma condio e criar mais de duas possibilidades de ao.<br>

Podemos escrever quantos `elif` quisermos, mas apenas um `if` e um `else` para analisar um conjunto de condies.<br>
        
Entendeu? Ento vamos explorar essa propriedade:<br>

```python
#considere agora as seguintes variveis
a = 10
b = 15

#dessa vez, tente criar diferentes condies e mudar os valores de 'a' e 'b'. Tente tambm adicionar elifs com diferentes
#condies mais especficas

if a < b:
    print('A primeira condio  verdadeira')
    
elif a > b:
    print('A segunda condio  verdadeira')
    
else:
    print('Nenhuma das condies  verdadeira')
```

Mas cuidado!!! Para tanto,  importante respeitar a indentao e os dois pontos que seguem essas expresses! Eles ditam quem est dentro de quem, dando sentido ao nosso programa. Vamos avaliar essa propriedade: <br>

Combinar operadores pode ficar muito complicado e, sem cuidado, podemos estar aplicando esses operadores com calores que no tinhamos inteno. Por isso, lembre-se de usar os parenteses para explicitar quais operaes voc est fazendo! Assim como na matemtica, as operaces entre parenteses acontecem primeiro!


Vamos ento praticar o aninhamento de outros operadores:

```python
#vamos considerar trs variveis
a = 10
b = 5
c = 17

#troque as comparaes, adicione 'ands' e 'ors', tente introduzir o termo 'not' e mude os valores de 'a', 'b' e 'c'
#dessa vez, troque tambem os parenteses, tente imaginar o que aconteceria com e resultado e compare com o resultado final

if not (a > b and a < c):
    print('O primeiro "if"  verdadeiro')

if (a < b or a < c) and (c > b):
    print('O segundo "if"  verdadeiro')
```

```python
#espere o professor/ a professora resolver a questo com voc


def maior(ns, n = -999999999):
    if not ns:
        return n  
    
    primeiro, resto = ns[0], ns[1:]
    print("Comparao: " ,n," > ", primeiro," ? ", end="")
    
    if n >= primeiro:
        print("Sim, continua com o maior atual: ", n)
        return maior(resto, n) 
    else:
        print("No, novo maior: ", primeiro)
        return maior(resto, primeiro)
    
maior([1,2,3,10,50,32,48,21])
```

Agora que voc j conhece os operadores condicionais e lgicos, agora aplicaremos um pouco desses conceitos no micro:bit.  impressionante o quanto esses operadores podem expandir seu leque de possibilidades!

Com o operador if, poderemos explorar muito mais os diversos recursos do micro:bit, dentre eles os `botes`. Cada um dos botes possui uma referncia, sendo o boto esquerdo chamado de 'A' e o boto direito de 'B'. Estes tambm possuem algumas funes, dentre elas destacamos:

`button_a.is_pressed()` -> A funo retorna Verdadeiro (caso o boto esteja pressionado) ou falso (caso o boto no esteja pressionado);<br>
`button_a.was_pressed()` -> A funo retorna Verdadeiro (caso o boto tenha sido pressionado) ou falso (caso o boto no tenha sido pressionado);<br>
`button_a.get_presses()` -> A funo retorna o nmero de vezes em que o boto foi pressionado.<br>

Repare que em todas as funes acima utilizamos "button_a", mas caso deseje utilizar o boto b, basta substituir por "button_b" e todas as funes funcionaro corretamente.<br>

Agora, vamos testar cada um desses comandos! Copie e complete os cdigos abaixo e execute-os em seu micro:bit.<br>

Primeiramente, testaremos o comando de manter o boto pressionado:

```python
from microbit import *

while True:
    #se o boto A est sendo pressionado 
        display.show(Image.SAD)      
    #caso contrrio
        display.show(Image.HAPPY)     
    sleep(1000)
```

O programa acima exibe uma cara triste no display caso voc esteja pressionando o boto A. Se ele no estiver pressionado, ser exibido o rosto feliz. Caso queira explorar um pouco mais o comando, voc pode acrescentar mais condicionais, mais um boto e diversas outras imagens para serem impressas no display!

Agora utilizaremos os dois botes e o comando que indica se os botes foram ou no pressionados:

```python
from microbit import *

while True:
    #se o boto A foi pressionado
        display.scroll("A DE AMOR")     
    #se o boto B foi pressionado
        display.show("B DE BAIXINHO")   
    sleep(1000)
```

Muito bem, voc j no  mais um baixinho! O programa exibe a frase "A DE AMOR" caso voc tenha pressionado o boto A e "B de baixinho" caso tenha pressionado o boto B.

Utilizando o comando que retorna quantas vezes o boto foi pressionado, podemos realizar algumas combinaes:

```python
from microbit import *

while True:
    #se o boto A foi pressionado mais de trs vezes
        display.scroll("RAPIDO DEMAIS!")    
    #se o boto B foi pressionado mais de cinco vezes
        display.scroll("VAI COM CALMA!")    
    sleep(10000)                            #intervalo de 10.000ms (10 segundos) para apertar os botes
```

So s cinco vezes em 10 segundos, OK?! Ao executar o cdigo acima, caso o boto A seja pressionado mais de 3 vezes, a frase "RAPIDO DEMAIS!"  mostrada no display. Se o boto B for pressionado mais de 5 vezes, a frase "VAI COM CALMA!" ser exibida.

J que voc conhece os trs comandos, complete o cdigo e conclua o programa abaixo: 

```python
while True:
    #se o boto A e o boto B estiverem pressionados
        display.show(Image.HEART)
        sleep(2000)
    #se o boto B foi pressionado
        display.show(Image.HAPPY)
    #se os botes foram pressionados 3 ou mais vezes
        display.show(Image.SAD)
        
    sleep(2000)
```
