## 1- Os operadores de comparação

No mundo da programação, um conceito muito útil e versátil é o da comparação. Em Python temos alguns operadores que conseguem realizar comparaçoes entre valores. Eles são:

`a > b    #maior que`<br>
`a < b    #menor que`<br>
`a >= b   #maior ou igual a`<br>
`a <= b   #menor ou igual a`<br>
`a == b   #igual a`<br>
`a != b   #diferente de`<br>

Com esses operadores podemos realizar diversas comparações com diversos tipos diferentes, comparando valores numéricos, strings, etc.Essas comparações são chamadas de expressões lógicas ou booleanas que são avaliadas como sendo verdadeiras ou falsas, em python essas expressões retornam um tipo Booleano com valor `True` ou `False`.

Vale resaltar que operador "=" se comporta como um operador de atribuição é não de igualdade como é o caso do "==".

Vamos observar os seguintes exemplos:

```python
#vamos considerar as seguintes variáveis
a = 10
b = 20.5

#agora vamos testar os nossos operadores de comparação. Troque os comparadores e veja qual o resultado!

a > b
```

Conseguiu? Provavelmente não, né? Isso é por que ainda falta uma peça nesse quebra-cabeças de comparação: Algo que consiga avaliar essa operação!<br>

Precisamos de uma maneira de pedir para o computador fazer coisas diferentes baseado na comparação entre os valores. Para isso usamos o termo `if`, que vem do inglês para `se`. Com esse termo, podemos pedir para o computador fazer `alguma coisa` `se` uma `condição` for atendida. Vamos tentar entender melhor:<br>

   Precisamos fazer com  que o computador<br>
    `imprima o valor 'a'` `se` `o valor 'a' for maior que o valor 'b'`<br>
     ou<br>
    `imprima o valor 'b'` `se` `o valor 'b' for maior que o valor 'a'`<br>
    
    
Entendeu? Agora vamos ver como escrever coisas desse tipo em um programa de verdade.<br>


Em Python utilizamos a expressão `if` da seguinte forma:<br>
    `if 'Alguma condição':
        'Alguma ação'`<br>

Perceba a indentação, pois ela é extremamente importante e precisa ser respeitada, assim como os dois pontos.<br>

A condição pode ser qualquer valor, e será considerada verdadeira para o booleano `True` e qualquer valor diferente de zero, nada (`None`) ou vazio, e falsa para o booleano `False` e qualquer um dos valores citados anteriormente. Vamos testar:


```python
#Vamos considerar a variável

a = True

#Troque o valor de 'a' e veja os casos em que seu valor é considerado verdadeiro
if a:
    print('A condição é verdadeira')

```

Além do `if` temos o operador `else`, que poderia ser traduzido para `caso contrário`, ou seja:<br>
    `if 'Alguma condição':
        'Alguma ação'
    else:
        'Alguma outra ação, caso a condição não seja atendida'`<br>

Podemos usar alguns diagramas para ajudar na compreenção do que acontece no programa quando usamos operadores condicionais:



Vamos testar essa propriedade: 


```python
#Novamente considere a variável

a = None


#Novamente, troque o valor de 'a' e perceba como conseguimos pedir para o computador fazer algo
#mesmo quando 'a' é falso

if a:
    print('Verdadeiro')

else:
    print('Falso')
```

Conseguiu? Eu sabia que conseguiria! Vamos então analisar mais uma propriedade de operadores condicionais: o `elif`<br>

O operador `elif` é uma contração do termo `else if` (se você já viu linguagens como C talvez conheça dessa forma) e significa `caso contrário, se`. Ele é muito útil para checar mais de uma condição e criar mais de duas possibilidades de ação.<br>

Vamos analisar um pouco de sua sintaxe:<br>
    `if 'Alguma condição':
        'Alguma ação'
    elif 'Alguma outra condição':
        'Alguma outra ação'      
    else:
        'Alguma outra ação, caso nenhuma das condições sejam atendidas'`<br>
        
Podemos escrever quantos `elif` quisermos, mas apenas um `if` e um `else` para analisar um conjunto de condições.<br>
        
Entendeu? Então vamos explorar essa propriedade:<br>

```python
#considere agora as seguintes variáveis
a = 10
b = 15

#dessa vez, tente criar diferentes condições e mudar os valores de 'a' e 'b'. Tente também adicionar elifs com diferentes
#condições mais específicas

if a < b:
    print('A primeira condição é verdadeira')
    
elif a > b:
    print('A segunda condição é verdadeira')
    
else:
    print('Nenhuma das condições é verdadeira')
```

Um último tema a ser abordado é o aninhamento de expressões `if`, `else` e `elif`:<br>

Podemos escrever expressões do tipo `if` dentro delas mesmas. Com isso, podemos criar condições dentro de condições, filtrando ainda mais nossos valores. Podemos olhar outro diagrama para ajudar na compreensão:



Mas cuidado!!! Para tanto, é importante respeitar a indentação e os dois pontos que seguem essas expressões! Eles ditam quem está dentro de quem, dando sentido ao nosso programa. Vamos avaliar essa propriedade: <br>



```python
#vamos levar em conta três variáveis
a = 1
b = 5
c = 0.5

#mais uma vez, mude as comparações, valores e tente acrescentar casos de 'elif'
if a <= b:
    if a > c:
        print('Primeiro "if" aninhado é verdadeiro')
    else:
        print('Primeiro "if" aninhado é falso')
else:
    print('Primeiro "if" é falso')
```

Esses operadores são muito úteis para avaliar mais de uma condição dentro do mesmo `if`. Podem ser traduzidos para `e` e `ou`, respectivamente, e podem ser utilizados da seguinte forma:<br>

Perceba como os operadores se comportam na presença de diferentes entradas. Agora vamos analisar essas propriedades um pouco melhor:

```python
#vamos considerar três variáveis
a = 10
b = 5
c = 17

#troque as comparações, adicione 'ands' e 'ors', tente introduzir o termo 'not' e mude os valores de 'a', 'b' e 'c'

if a > b and a < c:
    print('O primeiro "if" é verdadeiro')

if a < b or a < c:
    print('O segundo "if" é verdadeiro')
```

Podemos também aninhar os operadores lógicos para obtermos resultados diferentes nas operações. Vamos visualizar como isso aconteceria num exemplo simples: vamos combinar os operadores `and` e `not`! <br>

Se temos duas entradas verdadeiras, normalmente teriamos que:

|   A  |   B  | Saída |
|:----:|:----:|-------|
| true | true | true  |

Porém, se aplicarmos o operador `not` depois de aplicarmos o operador `and`, temos que a saída ficaria invertida:

|   A  |   B  | Saída |
|:----:|:----:|-------|
| true | true | false  |

Assim, analisando toda a tabela verdade desses operadores juntos, obtemos:

<table>
    <tr><th> <center>Operador AND com NOT</center> </th></tr>
<tr><td>

|   A   |   B   | Saída | 
|:-----:|:-----:|:-----:|
|  true | true  |  false |
|  true | false | true |
| false |  true | true |
| false | false | true |

</td></tr></table>

A ideia aqui é que os operadores fazem sua parte do trabalho separadamente, mas com os resultados de outros operadores. Combinar operadores pode ficar muito complicado e, sem cuidado, podemos estar aplicando esses operadores com calores que não tinhamos intenção. Por isso, lembre-se de usar os parenteses para explicitar quais operações você está fazendo! Assim como na matemática, as operacões entre parenteses acontecem primeiro!


Vamos então praticar o aninhamento de outros operadores:

```python
#vamos considerar três variáveis
a = 10
b = 5
c = 17

#troque as comparações, adicione 'ands' e 'ors', tente introduzir o termo 'not' e mude os valores de 'a', 'b' e 'c'
#dessa vez, troque tambem os parenteses, tente imaginar o que aconteceria com e resultado e compare com o resultado final

if not (a > b and a < c):
    print('O primeiro "if" é verdadeiro')

if (a < b or a < c) and (c > b):
    print('O segundo "if" é verdadeiro')
```

```python
#espere o professor/ a professora resolver a questão com você


def maior(ns, n = -999999999):
    if not ns:
        return n  
    
    primeiro, resto = ns[0], ns[1:]
    print("Comparação: " ,n," > ", primeiro," ? ", end="")
    
    if n >= primeiro:
        print("Sim, continua com o maior atual: ", n)
        return maior(resto, n) 
    else:
        print("Não, novo maior: ", primeiro)
        return maior(resto, primeiro)
    
maior([1,2,3,10,50,32,48,21])
```

## 4- Hora do micro:bit

Agora que você já conhece os operadores condicionais e lógicos, agora aplicaremos um pouco desses conceitos no micro:bit. É impressionante o quanto esses operadores podem expandir seu leque de possibilidades!

Com o operador if, poderemos explorar muito mais os diversos recursos do micro:bit, dentre eles os `botões`. Cada um dos botões possui uma referência, sendo o botão esquerdo chamado de 'A' e o botão direito de 'B'. Estes também possuem algumas funções, dentre elas destacamos:

`button_a.is_pressed()` -> A função retorna Verdadeiro (caso o botão esteja pressionado) ou falso (caso o botão não esteja pressionado);<br>
`button_a.was_pressed()` -> A função retorna Verdadeiro (caso o botão tenha sido pressionado) ou falso (caso o botão não tenha sido pressionado);<br>
`button_a.get_presses()` -> A função retorna o número de vezes em que o botão foi pressionado.<br>

Repare que em todas as funções acima utilizamos "button_a", mas caso deseje utilizar o botão b, basta substituir por "button_b" e todas as funções funcionarão corretamente.<br>

Agora, vamos testar cada um desses comandos! Copie e complete os códigos abaixo e execute-os em seu micro:bit.<br>

Primeiramente, testaremos o comando de manter o botão pressionado:

```python
from microbit import *

while True:
    #se o botão A está sendo pressionado 
        display.show(Image.SAD)      
    #caso contrário
        display.show(Image.HAPPY)     
    sleep(1000)
```

O programa acima exibe uma cara triste no display caso você esteja pressionando o botão A. Se ele não estiver pressionado, será exibido o rosto feliz. Caso queira explorar um pouco mais o comando, você pode acrescentar mais condicionais, mais um botão e diversas outras imagens para serem impressas no display!

Agora utilizaremos os dois botões e o comando que indica se os botões foram ou não pressionados:

```python
from microbit import *

while True:
    #se o botão A foi pressionado
        display.scroll("A DE AMOR")     
    #se o botão B foi pressionado
        display.show("B DE BAIXINHO")   
    sleep(1000)
```

Muito bem, você já não é mais um baixinho! O programa exibe a frase "A DE AMOR" caso você tenha pressionado o botão A e "B de baixinho" caso tenha pressionado o botão B.

Utilizando o comando que retorna quantas vezes o botão foi pressionado, podemos realizar algumas combinações:

```python
from microbit import *

while True:
    #se o botão A foi pressionado mais de três vezes
        display.scroll("RAPIDO DEMAIS!")    
    #se o botão B foi pressionado mais de cinco vezes
        display.scroll("VAI COM CALMA!")    
    sleep(10000)                            #intervalo de 10.000ms (10 segundos) para apertar os botões
```

São só cinco vezes em 10 segundos, OK?! Ao executar o código acima, caso o botão A seja pressionado mais de 3 vezes, a frase "RAPIDO DEMAIS!" é mostrada no display. Se o botão B for pressionado mais de 5 vezes, a frase "VAI COM CALMA!" será exibida.

Já que você conhece os três comandos, complete o código e conclua o programa abaixo: 

```python
while True:
    #se o botão A e o botão B estiverem pressionados
        display.show(Image.HEART)
        sleep(2000)
    #se o botão B foi pressionado
        display.show(Image.HAPPY)
    #se os botões foram pressionados 3 ou mais vezes
        display.show(Image.SAD)
        
    sleep(2000)
```

O seu programa deve mostrar o coração no display caso os dois botões estejam pressionados ao mesmo tempo, o rosto feliz caso o botão B tenha sido pressionado anteriormente e o rosto triste caso os dois tenham sido pressionados pelo menos 3 vezes.
