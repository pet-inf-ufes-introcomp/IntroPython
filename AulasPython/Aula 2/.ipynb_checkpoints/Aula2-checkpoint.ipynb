### 1. O que  um programa?

Um programa  uma sequncia de instrues que especificam como realizar uma tarefa. Essa tarefa pode ser algo como encontrar as razes de uma equao de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc. Normalmente essas instrues variam de acordo com a linguagem de programao que voc esta utilizando, mas  comum em praticamente todos os programas serem encontradas algumas instrues bsicas:

- Entrada: Receber informaes, seja do teclado, de um arquivo, do mouse, ou de um sensor;
- Sada: Enviar informaes, seja no terminal, num arquivo ou em um motor ou display;
- Matemtica: Realizar operaes matemticas normalmente  essencial para a maioria dos programas;
- Condicionais: Checar algumas condies e realizar a instruo apropriada;
- Repeties: Realizar alguma instruo repetidamente.

### 3. O que so e como criar variveis em Python

Como visto na Aula 1, variveis so espaos na memria destinados a armazenar informaes que sero utilizadas no futuro pelo computador.

As variveis podem ter qualquer nome, respeitando algumas regras:
    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, nmeros ou underscore
    
Normalmente, comeamos o nome da varivel com uma letra minuscula, e prximas palavras com letra maiscula. Alguns exemplos esto listados abaixo:

- dinheiroNaConta = 5000 
- idade_minima_permitida = 18
- cpfCliente = 123.456.789-10

 comum evitar acentos e outros caracteres especiais ao criar variveis!
    
Alm disso, existem algumas palavras reservadas na linguagem que no podem ser utilizadas como variveis, apesar de respeitarem as regras acima. Algumas dessas palavras so: True, False, continue, and, None, etc.

### A biblioteca math
J sabemos fazer matemtica bsica em Python!<br>

### 8. Como fazer o computador nos enviar informaes (sada)

Existem diversas maneiras de se imprimir as informaes que voc precisa. Ao contrrio do Jupyter, terminais (terminais tambm so chamados de "sada padro") e outros programas nem sempre imprimiro valores de variveis somente digitando o nome da varivel em uma linha. Alm disso, as vezes gostaramos de imprimir valores de variveis no meio de frases ou com formataes especficas. Para isso, utilizamos o comando `print()`.


```python
cafe_da_manha = "po com nutella e suco"
```

```python
print("quero " + cafe_da_manha + " no caf da manh")
```

```python
jogo_favorito = "Warcraft 3"
```

```python
print("hoje eu vou jogar", jogo_favorito, "com meus amigos")
```

Alm desses metodos bsicos de imprimir uma string, podemos fazer algumas coisas para ter mais controle e domnio sobre o que estamos printando. Primeiramente, vamos apresentar dois caracteres especiais, que sero muito teis no seu print:<br>

`"\n"` - representa uma quebra de linha, um "enter" no teclado<br><br>
`"\t"` - representa uma tabulao, um "tab" no teclado<br><br>

Aps colocar as vrgulas que separam as strings, podemos colocar algumas informaes adicionais, como:<br>

`sep = " "` - define o caractere que separa uma string da outra. ( um espao como padro)<br><br>
`end = "\n"` - define o caractere que sempre vai no final de um print. ( um "\n" como padro)<br><br>

```python
print("No caf da manh eu comi", cafe_da_manha, sep = " ")
```

```python
print("Faz alguns dias que no jogo", jogo_favorito, end = "\n")
```

Por fim, vamos aprender como inserir variveis no meio de uma string sem precisar separa-la com vrgulas ou operadores "+". Esta tcnica tambm  util para limitar o nmero de casas depois da vrgula de nmeros reais! Para colocarmos variveis no meio de uma string, primeiro precisamos conhecer os identificadores:<br>
`%d`<br>
`%f`<br>
`%s`<br>

Agora basta colocar os operadores no exato local que voc quer que a varivel aparea, e aps terminar a string, escreva `%nome_da_varivel`. Um exemplo deve tornar as coisas mais simples:<br>

```python
a = 22.5
print("Uma pessoa com o IMC %f est dentro do seu peso ideal!" %a)
```

```python
# bem fcil repetir o print que fizemos algumas clulas acima agora
print("quero %s no caf da manh" %cafe_da_manha)
```

Para limitar o nmero de casas decimais de uma varivel float, escrevemos a seguinte sintaxe:<br>
`%.xf`<br>
Onde x  o nmero de casas depois da vrgula que voc quer que o nmero tenha. Vamos ver o exemplo anterior novamente:

```python
print("Uma pessoa com o IMC %.2f est dentro do seu peso ideal!" %(a))
```

E se for preciso colocar mais de uma varivel na mesma string? Bom, a sintaxe mudar um pouco, mas  bem similar a anterior:<br>
Como sempre, um exemplo deve ilustrar melhor como fazer:<br>

```python
a = "farinha"
b = "ovos"
c = "acar"
d = "banana"
tempo = 40

print("Para fazer um bolo de banana, precisamos de %s, %s, %s e %s. Ele fica pronto em %d minutos!" %(a, b, c, d, tempo))
```

### 9. Como enviar informaes ao computador (entrada)
Imagine que voc fez um programa em Python que calcule a raiz quadrada da soma de dois nmeros. Agora, imagine que sempre que voc quiser utilizar esse programa voc tenha que abrir o arquivo em que ele est escrito e modificar os valores das variveis. Por sorte, existe uma maneira mais fcil de isso: utilizando o comando `input()`!

```python
x = input("Entre com uma frase: ")
```

```python
print('A frase "%s" no faz sentido!' %(x))
#perceba um dos porqus de aspas simples e duplas no fazerem diferena:fica bem mais fcil imprimir aspas no print
#alternativamente, voc pode printar as mesmas aspas do print, o caractere % e diversos outros colocando uma contra barra
#antes "\"
```

Repare que um input naturalmente cria uma string.
Para armazenar um inteiro ou um float, utilizamos `int(input())` e `float(input())` respectivamente.

```python
a = int(input("entre com um nmero inteiro: "))
b = float(input("entre com um nmero real: "))
print("O nmero inteiro  %d.\nO nmero real  %.2f\ne o resultado de %d/%.2f  %.2f" %(a, b, a, b, (a/b)))
```

### 10. Micro:bit Time

### Exerccio 1

Acessem a pgina https://python.microbit.org/v/1.1 e colem o cdigo abaixo <br>

`
from microbit import *
while True:<br>
    display.scroll('Introcomp!')
    sleep(2000)`
    
No cdigo existem algumas funes:

- A funo display.scroll() exibe a string informada entre parnteses deslizando-o pelos LEDs do microbit.
    - Existem outros parmetros que podem ser utilizados na funo display.scroll():
    - `display.scroll(string, delay = x, loop = False)`
    - sendo string a frase a ser impressa
    - x sendo o tempo de espera (em milissegundos)
    - e loop como uma booleana que diz se a impresso deve acontecer para sempre ou s uma vez
<br><br>
- A funo sleep() pausa a execuo do cdigo durante os milisegundos informados entre parnteses. Por exemplo, para esperar 2 segundos, precisamos colocar sleep(2000).

Agora modifique o cdigo acima criando uma string para armazenar seu nome e depois use a funo scroll para exibir a string nos LEDs
