### 1. O que é um programa?

Um programa é uma sequência de instruções que especificam como realizar uma tarefa. Essa tarefa pode ser algo como encontrar as raízes de uma equação de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc. Normalmente essas instruções variam de acordo com a linguagem de programação que você esta utilizando, mas é comum em praticamente todos os programas serem encontradas algumas instruções básicas:

- Entrada: Receber informações, seja do teclado, de um arquivo, do mouse, ou de um sensor;
- Saída: Enviar informações, seja no terminal, num arquivo ou em um motor ou display;
- Matemática: Realizar operações matemáticas normalmente é essencial para a maioria dos programas;
- Condicionais: Checar algumas condições e realizar a instrução apropriada;
- Repetições: Realizar alguma instrução repetidamente.

Acredite, qualquer programa que você fizer, por mais complicado que seja, é feito de instruções que se aproximam dessas 5 listadas acima. 
<br><br>
As pessoas se comunicam umas com as outras por meio de alguma linguagem. Podemos nos comunicar em português, inglês, francês, etc. Da mesma forma, precisamos de um jeito
para nos comunicar com o computador. Para isso, utilizamos alguma linguagem de programação, como C, C++, Java, PHP, Lua, etc.

### 3. O que são e como criar variáveis em Python

Como visto na Aula 1, variáveis são espaços na memória destinados a armazenar informações que serão utilizadas no futuro pelo computador.

As variáveis podem ter qualquer nome, respeitando algumas regras:
    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, números ou underscore
    
Normalmente, começamos o nome da variável com uma letra minuscula, e próximas palavras com letra maiúscula. Alguns exemplos estão listados abaixo:

- dinheiroNaConta = 5000 
- idade_minima_permitida = 18
- cpfCliente = 123.456.789-10

É comum evitar acentos e outros caracteres especiais ao criar variáveis!
    
Além disso, existem algumas palavras reservadas na linguagem que não podem ser utilizadas como variáveis, apesar de respeitarem as regras acima. Algumas dessas palavras são: True, False, continue, and, None, etc.

### A biblioteca math
Já sabemos fazer matemática básica em Python!<br>
Mas e se precisarmos fazer contas matemáticas mais complexas, como potenciação, radiciação, valores absolutos, senos e cossenos, etc?<br>Com sorte, Python tem uma biblioteca própria para isso, a biblioteca "math". Basta digitar `from math import *` no seu código, normalmente no começo do arquivo de código `.py` ou, se você estiver rodando o código em tempo de execução, digite antes de usar algo da biblioteca math pela primeira vez.
Faremos uma explicação mais detalhada sobre bibliotecas e funções no futuro.

Ao utilizar a biblioteca math, devemos nos atentar a alguns detalhes, como a sintaxe das funções matemáticas:<br>
para utilizar uma função, escrevemos: nome_da_funcao(valores)<br><br>
A seguir, alguns exemplos de funções da math:<br><br>
`ceil(valor)` - arredonda o número passado para cima<br><br>
`floor(valor)` - arredonda o número passado para baixo<br><br>
`abs(valor)` - retorna o módulo (valor absoluto) do número inteiro<br><br>
`factorial(valor)` - retorna o fatorial do número<br><br>
`pow(valor1, valor2)` - retorna o valor1 elevado ao valor2<br><br>
`sqrt(valor)` - retorna a raiz quadrada do valor<br><br>
`sin(valor), cos(valor), tan(valor)` - retorna, respectivamente, o seno, o cosseno e a tangente de um valor.<br>NOTA: o valor deve estar em radianos!<br><br>

Além disso, também temos as variáveis "pi" e "e", que também vem com a biblioteca. Por serem variáveis, você pode alterar seus valores do jeito que quiser, mas não é recomendado, uma vez que você pode voltar a precisar desses valores no futuro. Caso você tenha alterado essas variáveis e precise dos valores de novo, basta escrever: <br>
`pi = math.pi`<br>
`e = math.e`<br>

Caso você esteja curioso para ver o resto da documentação da biblioteca math, clique <a href = "https://docs.python.org/3/library/math.html">aqui</a>

### 7. Strings

Além de fazer operações matemáticas, é muito comum durante a criação de um programa dominar o uso de strings. A seguir, algumas técnicas para tratamento de strings em Python:

```python
#note que tanto aspas simples ('string') e aspas duplas ("string") fazem a mesma coisa!
frase = "isso é uma string"
```

```python
frase
```

### 8. Como fazer o computador nos enviar informações (saída)

Existem diversas maneiras de se imprimir as informações que você precisa. Ao contrário do Jupyter, terminais (terminais também são chamados de "saída padrão") e outros programas nem sempre imprimirão valores de variáveis somente digitando o nome da variável em uma linha. Além disso, as vezes gostaríamos de imprimir valores de variáveis no meio de frases ou com formatações específicas. Para isso, utilizamos o comando `print()`.


```python
cafe_da_manha = "pão com nutella e suco"
```

```python
print("quero " + cafe_da_manha + " no café da manhã")
```

```python
jogo_favorito = "Warcraft 3"
```

```python
print("hoje eu vou jogar", jogo_favorito, "com meus amigos")
```

Além desses metodos básicos de imprimir uma string, podemos fazer algumas coisas para ter mais controle e domínio sobre o que estamos printando. Primeiramente, vamos apresentar dois caracteres especiais, que serão muito úteis no seu print:<br>

`"\n"` - representa uma quebra de linha, um "enter" no teclado<br><br>
`"\t"` - representa uma tabulação, um "tab" no teclado<br><br>

Após colocar as vírgulas que separam as strings, podemos colocar algumas informações adicionais, como:<br>

`sep = " "` - define o caractere que separa uma string da outra. (é um espaço como padrão)<br><br>
`end = "\n"` - define o caractere que sempre vai no final de um print. (é um "\n" como padrão)<br><br>

```python
print("No café da manhã eu comi", cafe_da_manha, sep = " ")
```

```python
print("Faz alguns dias que não jogo", jogo_favorito, end = "\n")
```

Por fim, vamos aprender como inserir variáveis no meio de uma string sem precisar separa-la com vírgulas ou operadores "+". Esta técnica também é util para limitar o número de casas depois da vírgula de números reais! Para colocarmos variáveis no meio de uma string, primeiro precisamos conhecer os identificadores:<br>
`%d`<br>
`%f`<br>
`%s`<br>

Agora basta colocar os operadores no exato local que você quer que a variável apareça, e após terminar a string, escreva `%nome_da_variável`. Um exemplo deve tornar as coisas mais simples:<br>

```python
a = 22.5
print("Uma pessoa com o IMC %f está dentro do seu peso ideal!" %a)
```

```python
#é bem fácil repetir o print que fizemos algumas células acima agora
print("quero %s no café da manhã" %cafe_da_manha)
```

Para limitar o número de casas decimais de uma variável float, escrevemos a seguinte sintaxe:<br>
`%.xf`<br>
Onde x é o número de casas depois da vírgula que você quer que o número tenha. Vamos ver o exemplo anterior novamente:

```python
print("Uma pessoa com o IMC %.2f está dentro do seu peso ideal!" %(a))
```

E se for preciso colocar mais de uma variável na mesma string? Bom, a sintaxe mudará um pouco, mas é bem similar a anterior:<br>
Como sempre, um exemplo deve ilustrar melhor como fazer:<br>

```python
a = "farinha"
b = "ovos"
c = "açúcar"
d = "banana"
tempo = 40

print("Para fazer um bolo de banana, precisamos de %s, %s, %s e %s. Ele fica pronto em %d minutos!" %(a, b, c, d, tempo))
```

### 9. Como enviar informações ao computador (entrada)
Imagine que você fez um programa em Python que calcule a raiz quadrada da soma de dois números. Agora, imagine que sempre que você quiser utilizar esse programa você tenha que abrir o arquivo em que ele está escrito e modificar os valores das variáveis. Por sorte, existe uma maneira mais fácil de isso: utilizando o comando `input()`!

```python
x = input("Entre com uma frase: ")
```

```python
print('A frase "%s" não faz sentido!' %(x))
#perceba um dos porquês de aspas simples e duplas não fazerem diferença:fica bem mais fácil imprimir aspas no print
#alternativamente, você pode printar as mesmas aspas do print, o caractere % e diversos outros colocando uma contra barra
#antes "\"
```

Repare que um input naturalmente cria uma string.
Para armazenar um inteiro ou um float, utilizamos `int(input())` e `float(input())` respectivamente.

```python
a = int(input("entre com um número inteiro: "))
b = float(input("entre com um número real: "))
print("O número inteiro é %d.\nO número real é %.2f\ne o resultado de %d/%.2f é %.2f" %(a, b, a, b, (a/b)))
```

### Exercício 1

Acessem a página https://python.microbit.org/v/1.1 e colem o código abaixo <br>

`
from microbit import *
while True:<br>
    display.scroll('Introcomp!')
    sleep(2000)`
    
No código existem algumas funções:

- A função display.scroll() exibe a string informada entre parênteses deslizando-o pelos LEDs do microbit.
    - Existem outros parâmetros que podem ser utilizados na função display.scroll():
    - `display.scroll(string, delay = x, loop = False)`
    - sendo string a frase a ser impressa
    - x sendo o tempo de espera (em milissegundos)
    - e loop como uma booleana que diz se a impressão deve acontecer para sempre ou só uma vez
<br><br>
- A função sleep() pausa a execução do código durante os milisegundos informados entre parênteses. Por exemplo, para esperar 2 segundos, precisamos colocar sleep(2000).

Agora modifique o código acima criando uma string para armazenar seu nome e depois use a função scroll para exibir a string nos LEDs

### O que pode dar errado?

Erros acontecem com todo programador. Possuímos diferentes tipos de erros, alguns estão ligados a sintaxe e esses geralmente serão acusados pelo interpretador, o problema maior está naqueles com erros de semântica ou lógica que dependem totalmente do programador para encontrar e solucionar. Mas sempre tem uma forma de resolvê-los. 
Não desanimem e continuem programando!

Aqui estão alguns exemplos de erros comuns cometidos por programadores:

- Escrever algo da linguagem Python errado. Por exemplo, copie o trecho de código a seguir na celula abaixo e veja o que acontece: `pritn("Por que esse print não está funcionando? :c")`;
<br>

- Confundir letras maiúsculas e minúsculas nos nomes de variáveis: somaDosNumeros, SOMADOSNUMEROS, somadosnumeros são 3 variáveis diferentes;
<br>

- Indentação e espaços em branco. Siga um modelo de tabulação consistente e único no seu script. Se você começou usando dois espaços, ou tabulação de 4 espaços ou seja o que for, mantenha-a. O interpretador, geralmente, não saberá o que fazer se você misturar vários de tipos de tabulação e espaçamento.
    
Normalmente, ao ler os erros, é possível entender o problema, mas se você ainda não conseguiu consertar o erro, você pode consultar tanto um colega, um professor, ou até mesmo a internet!
