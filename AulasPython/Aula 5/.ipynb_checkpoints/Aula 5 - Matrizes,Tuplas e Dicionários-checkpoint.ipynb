## 1.1 O que é uma Matriz?

Matrizes são uma forma de organizar informações, aqui em python, uma matriz é na verdade uma lista de listas e é um tipo de estrutura de dados assim como uma lista. <br>
<br>Você se lembra de como uma lista é definida? Crie uma lista chamada __List__ de 5 elementos, onde esses elementos são valores númericos inteiros e a imprima.

Certo, se seu codigo funcionou, você deve ter printado sua lista de 5 elementos na tela. Mas como isso vai nos ajudar a compreender o que é uma matriz e seu funcionamento?

Vamos a um exemplo visual do que seria uma matriz.

### 1.2.2 Matriz de inputs
Já deu para se ter uma ideia de como declarar matrizes com valores estabelecidos pelo próprio programador. Agora vamos começar a declarar matrizes com inputs, ou seja, entradas estabelecidas durante a execução do programa. Antes de iniciar, vamos lembrar dos loops: a sintaxe do **for** é *for x in [1...10]*, onde o __x__ vai receber o valor de um elemento da lista a cada iteração do loop.
Uma função do Python que vai ser extremamente importante é a **append**. Essa função adiciona um elemento na ultima posição de uma lista. Por exemplo, se temos uma lista do tipo *List = [1,2,3,4]* e fizermos *List.append(5)*, teremos: *List = [1,2,3,4,5]*.

Vamos começar com a declaração de nossa matriz, acompanhe os códigos a seguir.
<br>
<br>Primeiro você declara uma lista vazia.

```python
#Aqui estamos criando uma lista vazia, que vai ser nossa base na criação da matriz.

A = []
```

Depois você lê o tamanho das linhas e das colunas da matriz a ser declarada (não esqueça de converter o tipo da leitura para *int* ).

```python
# Lendo o número de linhas e colunas
L = int(input("digite o tamanho das linhas:"))
C = int(input("digite o tamanho das colunas:"))
```

O código a seguir mostra a inicialização completa da matriz. Aqui temos que lembrar também da função range, que retorna uma lista: por exemplo range(5), nos retorna [0,1,2,3,4], isso nos possibilita percorrer a matriz de acordo com suas dimensões (L = linhas da matriz e C = colunas da matriz).

```python
#Toda a operação da criação da matriz
for l in range(L):
    #aqui vamos andando pelas linhas e criando uma lista vazia em cada linha
    A.append([])
    #caminhando pelas colunas da matriz
    for c in range(C):
        #aqui estamos declarando o elemento de cada posição da matriz, A[l][c],
        #onde a gente lê o que o usuario digitou e adiciona esse elemento na ultima posição da lista atual.
        A[l].append(int(input("digite um elemento da matriz: ")))
print(A)
```

O que esta acontecendo no código acima? No primeiro *for* nós estamos caminhando por cada elemento da lista L, que vai ser gerada com os elementos [ 0 , 1 , 2 , ... , L - 1 ]. A variável l vai assumir cada um dos valores da lista L, a cada iteração do loop. Por exemplo, no primeiro laço l vale 0, no segundo l vale 1 e assim por diante.

O A.appende( [ ] ), está criando uma lista vazia a cada laço no for.

Como explicado no primeiro *for*, o segundo caminha na lista gerada pelo *range*, onde c é cada valor da lista a cada laço.
<br><br>No *A[ l ].append( int ( input ( " digite um elemento da matrizz: " ) ) )*, o *input* vai pegar o valor digitado pelo usuario e o *append* vai adicionar esse valor na ultima posição da lista.
<br><br>O *A[ l ]* é uma lista, podemos interpretar como *A[ l ]* sendo a linha da matriz, l é o número da linha, e o elemento a ser adicionado vai ser adicionado na coluna c, por exemplo quando *c = 0*, iremos adicionar um elemento na posição *A[l][0]*, se *l = 1*, *então A[1][0]* é onde o elemento será adicionado.

# 2. Tuplas

## 2.1 O que é uma tupla?
Uma tupla é uma lista imutável. E o que quer dizer isso? Bem, nós trabalhamos com tuplas assim como trabalhamos com listas, a diferença é que quando definimos uma tupla não podemos mais editá-la. Com uma lista, sempre podemos ao longo do nosso programa adicionar ou remover elementos; já em uma tupla, não temos mais esse artificio.
<br> <br>
    Então podemos resumir uma tupla como sendo uma lista que tem as seguintes restrições:  
<br><br>
    - Não é possível adicionar novos elementos 
    - Não é possível remover elementos existentes 
    - Não é possível alterar os elementos 
    - Não é possível ordenar os elementos
<br><br>
   Mas, não se enganem, pois dizer que tuplas são somente listas imutáveis não está totalmente certo! Usamos isso para deixar o aprendizado dessa estrutura um pouco mais didático. Afinal as tuplas tem como objetivo trabalhar com dados heterogêneos. Por exemplo: <br>

## 2.4 Operações básicas com tuplas

As operações que fazemos com as listas também podem ser utilizadas com tuplas. Considere os exemplos a seguir:

```python
#aqui estamos criando uma tupla com os elementos 3 "vamos" 4  e "fugir"
t = (3,"vamos",4,"fugir")
#aqui estamos perguntando se 3 está dentro da nossa tupla, se sim retornara True se não retornara False.
3 in t
```

```python
#declaração de tupla
t1 = 2,"chega",555,"C++"
#concatenação das tuplas
t2 = t+t1 
#printando tupla resultante
print(t2)
```

```python
#tupla de somente 1 elemento
t3 = 1,
#concatenando essa tupla a ela mesma 5 vezes.
t4 = t3*5

print(t4)
```

Podemos acessar elementos da tupla como acessando elementos de uma lista.

```python
print(t)

print(t[0])
#somando o primeiro elemento com o terceiro elemento da tupla.
t5 = t[0]+t[2]

print(t5)

print(t[-1])

#o -1 é o indice do ultimo elemento da tupla, igual para lista.
```

Podemos fazer uma tupla de listas e também uma lista de tuplas, existe uma função chamada enumerate nativa da python capaz de nos retornar tuplas com os indices e os elementos dos indices da lista.


```python
l1 = ['a','b','c','d','e']
#criando lista de tuplas
l2 = list(enumerate(l1))
print(l2)
```

Se você observar verá que as tuplas seguem o padrão (indice, elemento) da lista para qual chamamos o enumerate.


```python
#criamos uma tupla de listas
tp = ([1,2,3],[4,5,6])
#printando o terceiro elemento da primeira posição da tupla
print(tp[0][2])
```


Outra operação que podemos fazer com elas são comparar se uma tupla é menor, maior ou igual.

```python
(1,2,3,4)<(2,3,4,5)
```

```python
#printando tupla recortada do segundo ao terceiro elemento
print(t[1:3])
```

```python
for x in ("a",1,2,3,"b",4,5,6,"c"):
    print(x)
```

# 3. Dicionários

*O que é um dicionário?* 

Quando você lê dicionário você pensa no livro de palavras e seus significados, certo? Pense que as palavras são "chaves" e os significados são "valores" associados a essas chaves. Para ser mais claro, o valor XXX.XXX.XXX-XX estaria associado a CPF, e quando fizessemos algo do tipo [ CPF ], nos retornariamos o valor XXX.XXX.XXX-XX. 

<br> Considere o exemplo da estrutura da figura abaixo. Na primeira coluna, temos as "chaves"; já na segunda coluna, tempos os "valores" associados a essas chaves. 
Em Python, podemos armazenar uma estrutura deste tipo em um dicionário. Um dicionário permite acesso rápido a valores e pode ser muito útil em programas!

Considere um exemplo de aplicação real de um dicionário: 
Imagine que você esteja construindo um sistema que armazena informações de uma pessoa, como por exemplo CPF, nome, idade, data de nascimento, profissão...
Concorda que é muito mais fácil de trabalhar com esses dados podendo acessá-los diretamente e de maneira eficiente?
Por exemplo, para acessar o nome da Pessoa, basta usar a chave adequada (por exemplo, "nome"), da seguinte forma: Pessoa [nome]

## 3.1 Declarando um Dicionário

O código a seguir mostra uma das maneiras de se declarar um dicionário, seguindo a sintaxe --> VARIAVEL = { 'CHAVE' : VALOR, 'CHAVE' : VALOR....}

```python
#declarando nosso dicionario pessoa
Pessoa1 = {'CPF':"034.894.345-87",'nome':"Matias Santos",'idade':43,'profissao':"Pedreiro",'DNascimento':"23/05/1098"}
print(Pessoa1)
print(Pessoa1['CPF'])
print(Pessoa1['nome'])
print(Pessoa1['idade'])
print(Pessoa1['profissao'])
print(Pessoa1['DNascimento'])
```

Curiosidade: Como acessamos os valores a partir de uma chave, não ordenamos dicionários, e a ordem em que adicionamos elementos a ele não importa!
<br><br>
Outra forma de se criar um dicionário é --> VARIAVEL = { }
<br>

VARIAVEL[ ' CHAVE ' ] = VALOR

```python
Pessoa2 = {}
Pessoa2['nome'] = "Geraldo"
Pessoa2['profissao'] = "Presidente"
Pessoa2['CPF'] = "942.231.241-12"
Pessoa2['DNascimento']="04/02/1292"
Pessoa2['idade']=423
print(Pessoa2)
```

```python
print("O" , Pessoa2['nome'] ,"é um otimo" , Pessoa2['profissao'] , "mas saiba que ele é um vampiro, ele tem tipo" , Pessoa2['idade'],"anos.")
```

Provamos que não precisamos adicionar elementos em ordem num dicionário. <br>

Curiosidade: podemos declarar um dicionário com um for ou while.  Por exemplo, se precisarmos criar um dicionário de tamanho 5 (5 chaves e 5 valores):

```python
#declarando dicionario vazio
dicionario = {}
#loop para criar chaves e associar a valores
for i in range(5):
    chave=input("digite a chave:")
    dicionario[chave]=input("digite o valor:")
print(dicionario)
```

## 3.2 Operações básicas com dicionários

Existem operações com dicionários que são bastante utéis quando se trabalha com essa estrutura. Vamos apresentar alguma delas aqui! <br>

*<h5>Como deletar um elemento de um dicionário</h5>*

A função _**del**_ recebe como parâmetro seu dicionário da forma --> NOME_DA_VARIAVEL[' CHAVE '], onde ele identifica o elemento que você quer deletar. 

```python
dicionario = {'nome':'João','profissao':'ferreiro','idade':'35'}
print('Antes:',dicionario)
print('-----------------------')
del dicionario['profissao']
print('Depois:',dicionario)
```

Deletamos do nosso dicionário o elemento associado a chave profissão, ou seja o conjunto *profissao : ferreiro*, foi retirado do dicionário.

*<h5>Assim como listas dicionários são mutáveis</h5>* 
<br>E dessa forma podemos mudar o valor de uma chave já existente.

```python
print('Antes:',dicionario)
print('-----------------------')
dicionario['nome'] = "Hodor"
print('Depois:',dicionario)
```

Obs: Podemos acumular valores já existentes no dicionário, por exemplo, se passou 10 anos desde que coletamos a informação do Hodor, sua idade agora seria a antiga + 10, podemos fazer isso da forma dicionario[' idade '] = dicionario [' idade '] + 10

*<h5>quantidade de elementos no dicionário</h5>*

Conseguimos saber a quantidade de elementos contidos em um dicionário a partir de uma função chamada len(dicionario). Como pode ser visto o parâmetro da função é o nome que você atribuiu a variável do tipo dicionário.

```python
print(len(dicionario))
```

*<h5>Iterando dicionários</h5>*<br>
Conseguimos andar sobre um dicionário com o for também, vamos printar os elementos do nosso dicionário como um exemplo.

```python
for i in dicionario:
    print(i)
```

```python
for x in dicionario.values():
    print (x)
```

```python
for chave in dicionario:
    print(chave,":",dicionario[chave])
```

*<h5>Os operadores in e not in</h5>* <br><br>
Também são validos nos dicionários, assim podemos verificar se um elemento está ou não contido no dicionário que estamos trabalhando.

```python
'profissao' in dicionario
```

```python
'CPF' in dicionario
```

```python
'profissao' not in dicionario
```

*<h5>copy()</h5>*
<br> Com essa função você copia um dicionário, dessa forma as alterações na copia não irão influenciar o original.

```python
fake = dicionario.copy()
print('Original:',dicionario)
print('Copia:',fake)
print('-----------------------')
del fake['nome']
print ('Copia sem nome:',fake)
print ('Original sem mudanças:',dicionario)
```

*<h5>Mesclando dicionários</h5>*
<br>
Existe a possibilidade de mesclar dois dicionários, fazemos isso com a função update(dicionario_de_masclagem).<br> Um exemplo de utilidade da mesclagem: Imagine que um grupo de pessoas estejam recolhendo informações separadas de um mesmo conjunto, e no final todas essas informações devem estar juntas.

```python
dic1 = {'cpf':"932.213.312-32",'nome':"Gutinho"}
dic2 = {'nomeDaMae':"Josephina",'nomeDoPai':"Vladimizinho"}
print('Dicionário 1:',dic1)
print('Dicionário 2:',dic2)
print('-----------------------')
dic1.update(dic2)
print('Dicionário Mesclado:',dic1)
```
