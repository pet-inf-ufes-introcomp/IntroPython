### 1.2.2 Matriz de inputs
J deu para se ter uma ideia de como declarar matrizes com valores estabelecidos pelo prprio programador. Agora vamos comear a declarar matrizes com inputs, ou seja, entradas estabelecidas durante a execuo do programa. Antes de iniciar, vamos lembrar dos loops: a sintaxe do **for**  *for x in [1...10]*, onde o __x__ vai receber o valor de um elemento da lista a cada iterao do loop.
Uma funo do Python que vai ser extremamente importante  a **append**. Essa funo adiciona um elemento na ultima posio de uma lista. Por exemplo, se temos uma lista do tipo *List = [1,2,3,4]* e fizermos *List.append(5)*, teremos: *List = [1,2,3,4,5]*.

Vamos comear com a declarao de nossa matriz, acompanhe os cdigos a seguir.
<br>
<br>Primeiro voc declara uma lista vazia.

```python
#Aqui estamos criando uma lista vazia, que vai ser nossa base na criao da matriz.

A = []
```

Depois voc l o tamanho das linhas e das colunas da matriz a ser declarada (no esquea de converter o tipo da leitura para *int* ).

```python
# Lendo o nmero de linhas e colunas
L = int(input("digite o tamanho das linhas:"))
C = int(input("digite o tamanho das colunas:"))
```

O cdigo a seguir mostra a inicializao completa da matriz. Aqui temos que lembrar tambm da funo range, que retorna uma lista: por exemplo range(5), nos retorna [0,1,2,3,4], isso nos possibilita percorrer a matriz de acordo com suas dimenses (L = linhas da matriz e C = colunas da matriz).

```python
#Toda a operao da criao da matriz
for l in range(L):
    #aqui vamos andando pelas linhas e criando uma lista vazia em cada linha
    A.append([])
    #caminhando pelas colunas da matriz
    for c in range(C):
        #aqui estamos declarando o elemento de cada posio da matriz, A[l][c],
        #onde a gente l o que o usuario digitou e adiciona esse elemento na ultima posio da lista atual.
        A[l].append(int(input("digite um elemento da matriz: ")))
print(A)
```

O que esta acontecendo no cdigo acima? No primeiro *for* ns estamos caminhando por cada elemento da lista L, que vai ser gerada com os elementos [ 0 , 1 , 2 , ... , L - 1 ]. A varivel l vai assumir cada um dos valores da lista L, a cada iterao do loop. Por exemplo, no primeiro lao l vale 0, no segundo l vale 1 e assim por diante.

O A.appende( [ ] ), est criando uma lista vazia a cada lao no for.

Como explicado no primeiro *for*, o segundo caminha na lista gerada pelo *range*, onde c  cada valor da lista a cada lao.
<br><br>No *A[ l ].append( int ( input ( " digite um elemento da matrizz: " ) ) )*, o *input* vai pegar o valor digitado pelo usuario e o *append* vai adicionar esse valor na ultima posio da lista.
<br><br>O *A[ l ]*  uma lista, podemos interpretar como *A[ l ]* sendo a linha da matriz, l  o nmero da linha, e o elemento a ser adicionado vai ser adicionado na coluna c, por exemplo quando *c = 0*, iremos adicionar um elemento na posio *A[l][0]*, se *l = 1*, *ento A[1][0]*  onde o elemento ser adicionado.

Para ser mais claro, o valor XXX.XXX.XXX-XX estaria associado a CPF, e quando fizessemos algo do tipo [ CPF ], nos retornariamos o valor XXX.XXX.XXX-XX. 

<br> Considere o exemplo da estrutura da figura abaixo. Na primeira coluna, temos as "chaves"; j na segunda coluna, tempos os "valores" associados a essas chaves. 
Em Python, podemos armazenar uma estrutura deste tipo em um dicionrio. Um dicionrio permite acesso rpido a valores e pode ser muito til em programas!

Considere um exemplo de aplicao real de um dicionrio: 
Imagine que voc esteja construindo um sistema que armazena informaes de uma pessoa, como por exemplo CPF, nome, idade, data de nascimento, profisso...
Concorda que  muito mais fcil de trabalhar com esses dados podendo acess-los diretamente e de maneira eficiente?
Por exemplo, para acessar o nome da Pessoa, basta usar a chave adequada (por exemplo, "nome"), da seguinte forma: Pessoa [nome]

## 3.1 Declarando um Dicionrio

O cdigo a seguir mostra uma das maneiras de se declarar um dicionrio, seguindo a sintaxe --> VARIAVEL = { 'CHAVE' : VALOR, 'CHAVE' : VALOR....}

```python
#declarando nosso dicionario pessoa
Pessoa1 = {'CPF':"034.894.345-87",'nome':"Matias Santos",'idade':43,'profissao':"Pedreiro",'DNascimento':"23/05/1098"}
print(Pessoa1)
print(Pessoa1['CPF'])
print(Pessoa1['nome'])
print(Pessoa1['idade'])
print(Pessoa1['profissao'])
print(Pessoa1['DNascimento'])
```

Curiosidade: Como acessamos os valores a partir de uma chave, no ordenamos dicionrios, e a ordem em que adicionamos elementos a ele no importa!
<br><br>
Outra forma de se criar um dicionrio  --> VARIAVEL = { }
<br>

VARIAVEL[ ' CHAVE ' ] = VALOR

```python
Pessoa2 = {}
Pessoa2['nome'] = "Geraldo"
Pessoa2['profissao'] = "Presidente"
Pessoa2['CPF'] = "942.231.241-12"
Pessoa2['DNascimento']="04/02/1292"
Pessoa2['idade']=423
print(Pessoa2)
```

```python
print("O" , Pessoa2['nome'] ," um otimo" , Pessoa2['profissao'] , "mas saiba que ele  um vampiro, ele tem tipo" , Pessoa2['idade'],"anos.")
```

Provamos que no precisamos adicionar elementos em ordem num dicionrio. <br>

Curiosidade: podemos declarar um dicionrio com um for ou while.  Por exemplo, se precisarmos criar um dicionrio de tamanho 5 (5 chaves e 5 valores):

```python
#declarando dicionario vazio
dicionario = {}
#loop para criar chaves e associar a valores
for i in range(5):
    chave=input("digite a chave:")
    dicionario[chave]=input("digite o valor:")
print(dicionario)
```

## 3.2 Operaes bsicas com dicionrios

Existem operaes com dicionrios que so bastante utis quando se trabalha com essa estrutura. Vamos apresentar alguma delas aqui! <br>

*<h5>Como deletar um elemento de um dicionrio</h5>*

A funo _**del**_ recebe como parmetro seu dicionrio da forma --> NOME_DA_VARIAVEL[' CHAVE '], onde ele identifica o elemento que voc quer deletar. 

```python
dicionario = {'nome':'Joo','profissao':'ferreiro','idade':'35'}
print('Antes:',dicionario)
print('-----------------------')
del dicionario['profissao']
print('Depois:',dicionario)
```

Deletamos do nosso dicionrio o elemento associado a chave profisso, ou seja o conjunto *profissao : ferreiro*, foi retirado do dicionrio.

*<h5>Assim como listas dicionrios so mutveis</h5>* 
<br>E dessa forma podemos mudar o valor de uma chave j existente.

```python
print('Antes:',dicionario)
print('-----------------------')
dicionario['nome'] = "Hodor"
print('Depois:',dicionario)
```

Obs: Podemos acumular valores j existentes no dicionrio, por exemplo, se passou 10 anos desde que coletamos a informao do Hodor, sua idade agora seria a antiga + 10, podemos fazer isso da forma dicionario[' idade '] = dicionario [' idade '] + 10

*<h5>quantidade de elementos no dicionrio</h5>*

Conseguimos saber a quantidade de elementos contidos em um dicionrio a partir de uma funo chamada len(dicionario). Como pode ser visto o parmetro da funo  o nome que voc atribuiu a varivel do tipo dicionrio.

```python
print(len(dicionario))
```

Conseguimos andar sobre um dicionrio com o for tambm, vamos printar os elementos do nosso dicionrio como um exemplo.

```python
for i in dicionario:
    print(i)
```

*<h5>copy()</h5>*
<br> Com essa funo voc copia um dicionrio, dessa forma as alteraes na copia no iro influenciar o original.

```python
fake = dicionario.copy()
print('Original:',dicionario)
print('Copia:',fake)
print('-----------------------')
del fake['nome']
print ('Copia sem nome:',fake)
print ('Original sem mudanas:',dicionario)
```

*<h5>Mesclando dicionrios</h5>*
<br>
Existe a possibilidade de mesclar dois dicionrios, fazemos isso com a funo update(dicionario_de_masclagem).<br> Um exemplo de utilidade da mesclagem: Imagine que um grupo de pessoas estejam recolhendo informaes separadas de um mesmo conjunto, e no final todas essas informaes devem estar juntas.

```python
dic1 = {'cpf':"932.213.312-32",'nome':"Gutinho"}
dic2 = {'nomeDaMae':"Josephina",'nomeDoPai':"Vladimizinho"}
print('Dicionrio 1:',dic1)
print('Dicionrio 2:',dic2)
print('-----------------------')
dic1.update(dic2)
print('Dicionrio Mesclado:',dic1)
```
