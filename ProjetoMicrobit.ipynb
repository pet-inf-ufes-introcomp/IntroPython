## Uma base teórica...

O micro:bit possui diversos pinos, e nessa aula utilizaremos quatro deles. Mas como funcionam esses pinos?

<img src="https://cdn.multilogica-shop.com/microbit-front.png" alt="Pinos" style="width:300px; height:250px"/>

Antes de qualquer coisa, você deve ter notado que temos pinos numerados de 0 a 2 e também temos um pino diferente, escrito GND. GND é a abreviação para a palavra "ground", traduzindo diretamente para o português, "terra". O GND é a parte negativa do seu circuito!

Circuito? Mas como assim? O micro:bit em seu padrão funciona como um circuito aberto em cada pino, sendo os numerados a parte positiva do seu circuito e o GND é a parte negativa de todos eles. Como é aberto, não passa corrente entre eles.

Lembra da primeira aula, quando você fez seu primeiro código em blocos e uma das funcionalidades era utilizar os pinos? Muitas vezes ocorriam problemas pois, apesar do código corresponder à ideia de apertar o pino, o usuário esquece de apertar também o GND. Ao apertar um dos pinos numerados e o GND, seu circuito se torna fechado, ou seja, passa corrente do pino numerado até o GND. O micro:bit detecta justamente essa passagem de corrente para realizar as ações requisitadas pelo código.

##### Então isso quer dizer que eu vou tomar um CHOQUE??? 
Sim e não. Realmente passará uma corrente pelo seu corpo até chegar no GND, mas o valor é mínimo, imperceptível!

O jogo deve funcionar até que o jogador aperte o botão errado. Enquanto os botões forem apertados corretamente, o jogo deve continuar gerando números aleatórios, e cada vez mais rápido, aumentando a chance de erro.

A velocidade de reação de uma pessoa saudável varia de 0,15 até 0,45 segundos. Esse valor ainda não considera o tempo em que o jogador demora para deslocar o membro de um botão pro outro. Para contornarmos essa situação, é recomendável que se estabeleça um limite de decaimento da variável tempo (um valor mínimo entre loops). O tempo que utilizarei será de 660 milissegundos.

## 5.1 Múltiplos Botões

E se eu resolver apertar todos os botões ao mesmo tempo? Pensando bem na parte elétrica, isso funcionaria no nosso código atual. **Vamos solucionar esse problema imediatamente antes que alguém descubra e bata meu recorde!**

A maneira escolhida para contornar esse problema é realizar checagens durante o tempo de decisão do jogador. A quantidade de checagens fica a escolha do programador. Um valor entre 10 e 30 é conveniente para evitar trapaças do jogador e também não exigir muito processamento do Micro:bit. Mas como faremos isso?

Com auxílio da matemática básica, faremos um loop com 20 iterações e um intervalo de avaliação dentro dele. Como já temos uma variével chamada tempo com o intervalo, basta utilizarmos ele dentro desse loop, porém dividindo por 20 (exatamente o número de iterações). Criaremos também uma variável **erro** que, caso tenha seu valor alterado, futuramente interromperá o loop e consequentemente ocasiona na derrota do jogador.

A implementação da função multiplosBotoes() é bem simples, realizando a checagem da combinação dois a dois dos pinos, ou seja, verificando, em todas as combinações possíveis, se ao menos dois dos três pinos estão pressionados ao mesmo tempo.

```python
def multiplosBotoes():
    if (pin0.is_touched() and pin1.is_touched()) or (pin0.is_touched() and pin2.is_touched()) or (pin1.is_touched() and pin2.is_touched()):
        return True
    else:
        return False
```

## 5.2 Troca de Botões
Basta um tempo mínimo de jogatina para perceber outra falha: caso você aperte o botão errado e retorne para o certo durante o intervalo de um loop para o outro, o jogo desconsidera sua falha. Isso ocorre pois o único botão considerado pelo jogo é o botão que está apertado exatamente no momento em que o intervalo de tempo termina.

Para alterarmos esse problema, devemos fazer mais uma função, desta vez verificando se o botão apertado condiz ou não com o número exibido no display.

```python
def botaoTrocado(numero):
    if numero == 0 and pin1.is_touched():
        return True
    elif numero == 0 and pin2.is_touched():
        return True
    elif numero == 1 and pin0.is_touched():
        return True
    #Outras verificações...
    else
        return False
```

Aproveitando o loop que fizemos para verificar a função multiplosBotoes(), acrescentaremos a nova função com um operador lógico or:

```python
#Parte do código foi ocultado para evitar confusão e desfoque

erro = 0

while True:

    for i in range(10):
        if multiplosBotoes() or botaoTrocado(aleatorio):
            erro += 1
            break
        sleep(tempo/10)
    
    if botaoCerto(aleatorio):
        display.show(Image.YES)
        
    else:
        erro += 1
        
    if erro > 0:
        #Encerra o jogo
```

Com um olhar mais aguçado, percebemos uma peculiaridade entre as duas funções: **a função botaoTrocado() já supre os retornos lógicos de multiplosBotoes(). Isso quer dizer que, se removermos a função multiplosBotoes() da condicional, o jogo funcionará da mesma maneira!**

Interpretando a partir da função botaoTrocado(), basta lembrarmos que ela faz a verificação do número e do botão apertado, checando a compatibilidade entre eles. Como o número corresponde a um e somente um botão, ao apertar dois ou mais simultaneamente a nossa função acusa o erro do jogador.

Aqui cabe uma reflexão sobre o código: eu deveria ou não fazer a função botaoTrocado()?
Apesar do gasto computacional durante as verificações, em projetos colaborativos de maior escala é indispensável a legibilidade do código. Ao atribuir "duas funções" para uma função, isso tudo pode se confundir futuramente em uma atualização ou no reuso, trazendo grandes problemas... **Modularização é sempre bem-vinda!**
